## Computer -  Rock Paper Scissors

<font face = "Tahoma" , size=3> Setup

```python
#list_human = ["Rock","Paper","Scissors","Quit"]
#dict_human = {0:"rock",1:"paper",2:"scissors"}

human_input_list = []
comp_input = []
game_results  = {"win":0,"loss":0,"tie":0}
game_outputs =  {"rock":0,"paper":0,"scissor":0,"quit":0}
```

```python
def keywithmaxval(d):
    v=list(d.values())
    k=list(d.keys())
    v2 = sorted(v)
    if(v2[-1] == v2[-2] & v2[-2]==v2[-3]):
        output_k = random.choice(k)
    elif(v2[-1] == v2[-2]):
        max_value = max(v)  
        output_k = random.choice(list({key for key, value in d.items() if value == max_value}))
    else:
        output_k = k[v.index(max(v))]   
    return output_k

def  compOutput(d):
    comp_output = "None"
    if(keywithmaxval(d) == "rock"):
        comp_output = "paper"
    if(keywithmaxval(d) == "paper"):
        comp_output = "scissor"
    if(keywithmaxval(d) == "scissor"):
        comp_output = "rock"
    return comp_output
```

<font face = "Tahoma" , size=3> User Input

```python
while True:
    try:
        Human_input = str(input("Please enter user choice : \n"))
    except ValueError:
        clear()
        print("Incorrect format. Please try again with correct choice\n")
        continue
    if not Human_input.lower() in game_outputs.keys():
        print("Not valid choice. Please try again\n")
        continue
    if Human_input.lower() == "quit":
        print(len(human_input_list))
        print(game_results["win"])
        break
    else:
        Human_input =Human_input.lower()
        human_input_list.append(Human_input)
        comp_input.append(compOutput(game_outputs))
        if(human_input_list[-1] == comp_input[-1]):
            print("Tie\n")
            game_results["tie"]+=1
            game_outputs[Human_input]+=1
        elif(((human_input_list[-1] == "rock") & (comp_input[-1] == "paper")) | ((human_input_list[-1] == "paper") & (comp_input[-1] == "scissor")) | ((human_input_list[-1] == "scissor") & (comp_input[-1] == "rock"))):
            print("User loses\n")
            game_results["loss"]+=1
            game_outputs[Human_input]+=1
        else:
            print("User wins\n")
            game_results["win"]+=1
            game_outputs[Human_input]+=1
        continue
```

```python
def moving_window(tw):
    new_dict = {}
    tw_cleaned = re.sub(r"\s[^\w]+\s"," ",str(tw))
    split_list  = tw_cleaned.strip().split(' ')
    #print(split_list)
    split_list_2 =[y for y in [split_list[i:i+3] for i in range(0, len(split_list)-2)]]
    #print(split_list_2)
    split_list_2_joined = [' '.join(element) for element in split_list_2]
    new_dict = {key:value for key, value in zip(list(set(split_list_2_joined)),[1]*len(list(set(split_list_2_joined))))}   
    return new_dict 
```

```python
import math
def cosine(dict1=moving_window(k[1]),dict2=moving_window(k[3])):
    numerator = 0
    denominator = 0
    for i in  dict1.keys():
        if i in dict2.keys():
            numerator+=1
    n1 = len(list(dict1.keys()))
    n2 = len(list(dict2.keys()))
    denominator = math.sqrt(n1*n2)
    return  numerator/denominator 
```

```python
#Have list for keys, Need list of dictionaries as values
def Frequency(quote_list_of_words):
    # Creating an empty dictionary
    freq_dict = {key:value for key,value in zip(list(set(quote_list_of_words)),[0]*len(list(set(quote_list_of_words))))}
    for item in quote_list_of_words:
        if (item in freq_dict):
            freq_dict[item] += 1
        else:
            freq_dict[item] = 1
    return freq_dict
Frequency(quote_words(quotes_list_cleaned[1]))
```

<font face="Tahoma" size=3> Quote  search  using  multiple  words

```python
# Add two dicts
def addDicts(dict1,dict2):
    d3 = dict(dict1)
    d3.update(dict2) 
    for i, j in dict1.items():
        for x, y in dict2.items():
            if i == x:
                d3[i]=(j+y)
    return(d3)
```

```python
def quote_search_multiple(word_list):
    list_of_dictionaries = [quote_search(element) for element in word_list]
    temp_dict = {}
    for i in range(len(list_of_dictionaries)):
        temp_dict = addDicts(temp_dict,list_of_dictionaries[i])
    return(temp_dict)
```
